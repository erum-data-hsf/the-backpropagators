# Activation Function Practice

This notebook is designed to present hands-on exercises for activation functions. By implementing various activation functions from scratch, you will gain a deeper understanding of their mechanisms and practical effects.

```python
## import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

## ReLU (Rectified Linear Unit)

The ReLU activation function outputs the input directly if it is positive; otherwise, it outputs zero. It is widely used in deep learning because it helps models converge faster and reduces the likelihood of vanishing gradients. The function is defined as:

$$
y = 
\begin{cases}
x, & \text{if } x \geq 0 \\
0, & \text{if } x < 0
\end{cases}
$$

# Exercise 1: 
Implement the ReLU activation function by hand.  
- Write a Python function that takes a number or an array as input and returns the ReLU output.
- Plot the function curve for inputs ranging from -10 to 10.
- Compare your result with PyTorch's built-in ReLU implementation.

```python
# ReLU activation function practice (hand-written)
def relu(x):
    return np.maximum(0, x) # Code ReLU function here

x = np.linspace(-10, 10, 400)
y = relu(x)

x_torch = torch.linspace(-10, 10, 400)
relu_layer = nn.ReLU()
y_torch = relu_layer(x_torch)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Hand-written ReLU
axs[0].plot(x, y, label='Hand-written ReLU', color='blue')
axs[0].set_title('ReLU Activation Function (Practice)')
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')
axs[0].axhline(0, color='black', linewidth=0.5, ls='--')
axs[0].legend()

# PyTorch ReLU
axs[1].plot(x_torch.numpy(), y_torch.numpy(), label='PyTorch ReLU', color='green')
axs[1].set_title('ReLU Activation Function (PyTorch)')
axs[1].set_xlabel('Input')
axs[1].set_ylabel('Output')
axs[1].axhline(0, color='black', linewidth=0.5, ls='--')
axs[1].legend()
```

## Leaky ReLU

The Leaky ReLU activation function is a variant of ReLU that allows a small, non-zero gradient when the input is negative. This helps to prevent the "dying ReLU" problem, where neurons become inactive and only output zero. The function is defined as:

$$
y = 
\begin{cases}
x, & \text{if } x \geq 0 \\
\alpha x, & \text{if } x < 0
\end{cases}
$$

where $\alpha$ is a small positive constant (commonly 0.01).

# Exercise 2:
Implement the Leaky ReLU activation function by hand.  
- Write a Python function that takes a number or an array as input and returns the Leaky ReLU output.
- Plot the function curve for inputs ranging from -10 to 10.
- Compare your result with PyTorch's built-in LeakyReLU implementation.

```python
# Leaky ReLU activation function practice (hand-written)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x) # Code Leaky ReLU function here

x = np.linspace(-10, 10, 400)
y = leaky_relu(x)

x_torch = torch.linspace(-10, 10, 400)
leaky_relu_layer = nn.LeakyReLU(negative_slope=0.01)
y_torch = leaky_relu_layer(x_torch)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
# Hand-written Leaky ReLU
axs[0].plot(x, y, label='Hand-written Leaky ReLU', color='blue')
axs[0].set_title('Leaky ReLU Activation Function (Practice)')
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')
axs[0].axhline(0, color='black', linewidth=0.5, ls='--')
axs[0].legend()
# PyTorch Leaky ReLU
axs[1].plot(x_torch.numpy(), y_torch.numpy(), label='PyTorch Leaky ReLU', color='green')
axs[1].set_title('Leaky ReLU Activation Function (PyTorch)')
axs[1].set_xlabel('Input')
axs[1].set_ylabel('Output')
axs[1].axhline(0, color='black', linewidth=0.5, ls='--')
axs[1].legend()
plt.show()  
```

## Softmax

The Softmax activation function is commonly used in the output layer of classification models. It converts a vector of raw scores (logits) into probabilities, where each value is in the range (0, 1) and the sum of all values is 1. This makes it suitable for multi-class classification tasks.

The function is defined as:

$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$

where $x_i$ is the $i$-th element of the input vector $x$.

# Exercise 3:
Implement the Softmax activation function by hand.  
- Write a Python function that takes a vector as input and returns the Softmax output.
- Test your function with a sample input vector.
- Compare your result with PyTorch's built-in Softmax implementation.

```python
# Sigmoid activation function practice (hand-written)
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) # Code Sigmoid function here

x = np.linspace(-10, 10, 400)
y = sigmoid(x)

x_torch = torch.linspace(-10, 10, 400)
sigmoid_layer = nn.Sigmoid()
y_torch = sigmoid_layer(x_torch)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
# Hand-written Sigmoid
axs[0].plot(x, y, label='Hand-written Sigmoid', color='blue')
axs[0].set_title('Sigmoid Activation Function (Practice)')
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')
axs[0].axhline(0, color='black', linewidth=0.5, ls='--')
axs[0].legend()
# PyTorch Sigmoid
axs[1].plot(x_torch.numpy(), y_torch.numpy(), label='PyTorch Sigmoid', color='green')
axs[1].set_title('Sigmoid Activation Function (PyTorch)')
axs[1].set_xlabel('Input')
axs[1].set_ylabel('Output')
axs[1].axhline(0, color='black', linewidth=0.5, ls='--')
axs[1].legend()
plt.show()
```
